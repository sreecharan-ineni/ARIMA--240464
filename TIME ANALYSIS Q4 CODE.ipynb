```python
import yfinance as yf
import pandas as pd

data = yf.download('AAPL', start='2019-01-01', end='2023-12-31')
data = data[['Open', 'High', 'Low', 'Close', 'Volume']]
data.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(data['Close'], label='AAPL Closing Price', color='blue')
plt.title('AAPL Stock Price (2019â€“2023)')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(data['Close'], model='multiplicative', period=252)  # Approx. number of trading days per year

decomposition.plot()
plt.tight_layout()
plt.show()
```
