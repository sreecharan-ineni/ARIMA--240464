```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sales = np.array([20, 24, 22, 26, 25, 28, 30, 27, 29, 31])
months = pd.date_range(start='2023-01', periods=len(sales), freq='M')

df = pd.DataFrame({'Month': months, 'Sales': sales})
df.set_index('Month', inplace=True)

df['3_MA'] = df['Sales'].rolling(window=3, center=True).mean()

plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Month', y='Sales', label='Original Sales', marker='o')
sns.lineplot(data=df, x='Month', y='3_MA', label='3-Point Moving Average', marker='s')
plt.title('Monthly Sales with 3-Point Moving Average')
plt.xlabel('Month')
plt.ylabel('Sales (in thousands)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
